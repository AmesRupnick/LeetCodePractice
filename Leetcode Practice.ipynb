```python
def minSubArrayLen(target, nums):
    """
    :type target: int
    :type nums: List[int]
    :rtype: int
    find the size of the smallest sub array that adds up to the target
    """
    windowSum = 0
    wSize = 1
    arrSize = len(nums)
    check = False
    while True:
        i = 0
        for i in range(wSize):
            windowSum = windowSum + nums[i]
        i = wSize
        if windowSum >= target:
            return wSize
        while i < arrSize:
            #instead of adding each time since that will slow things down, subtract tail and add head
            windowSum = windowSum - nums[i - wSize]
            windowSum = windowSum + nums[i] 
            if windowSum >= target:
                check = True
            i = i + 1
        if check == True:
            return wSize
        else:
            wSize = wSize + 1
        windowSum = 0 
        if wSize > arrSize:
            return 0

nums = [1, 1, 1, 1] #issue here still
target = 5
length = minSubArrayLen(target, nums)
print(f"Minimum length is {length}.")
```

```python
def isIsomorphic(s, t):
    """
    :type s: str
    :type t: str
    :rtype: bool
    use a hash map to check if the word is isomorphic
    """
    s_hash = 200*[0] #instead, map the letter of t to the array s and check to see if correct
    t_hash = 200*[0]
    for i in range (len(s)):
        if s_hash[ord(s[i])] == 0:
            if t_hash[ord(t[i])] == 0:
                s_hash[ord(s[i])] = ord(t[i])
                t_hash[ord(t[i])] = ord(s[i])
            else:
                return False
        else:
            if s_hash[ord(s[i])] != ord(t[i]):
                return False
    return True
    

s = "egg"
t = "adl"
print(isIsomorphic(s, t))
```

```python
def summaryRanges(nums):
        """
        :type nums: List[int]
        :rtype: List[str]
        use intervals to summarize the ranges in an ordered list
        """
        intervals = []
        interval = ""
        i = 0
        if len(nums) == 1:
            full_interval = str(nums[0])
            return [full_interval]
        for x in nums:
            if i == 0:
                interval = x
            elif x != nums[i - 1] + 1: 
                if nums[i - 1] == interval:
                    full_interval = str(interval)
                    intervals.append(full_interval)
                else:
                    interval_start = str(interval)
                    interval_end = str(nums[i - 1])
                    full_interval = interval_start+"->"+interval_end
                    intervals.append(full_interval)
                interval = x
                if i == len(nums) - 1:
                    full_interval = str(interval)
                    intervals.append(full_interval)
                    continue
            if i == len(nums) - 1:
                interval_start = str(interval)
                interval_end = str(x)
                full_interval = interval_start+"->"+interval_end
                intervals.append(full_interval)
            i = i + 1 
        return intervals

nums = [-1]
intervals = summaryRanges(nums)
print(intervals)
```

```python
def isValid(s):
    """
    :type s: str
    :rtype: bool
    determine if parethesis are correctly matched
    failed, check https://leetcode.com/problems/valid-parentheses/solutions/4672296/stack-approach/?envType=study-plan-v2&envId=top-interview-150
    """ 
    length = len(s)
    checkArray = []
    if length%2 == 1:
        return False
    for i in range(length):    
        if s[i] == "(" or s[i] == "[" or s[i] == "{":
            checkArray.append(s[i])
        else:
            if len(checkArray) == 0:
                return False
            if s[i] == ")":
                if checkArray[len(checkArray)-1] == "(":
                    del checkArray[len(checkArray)-1]
                else:
                    return False
            if s[i] == "]":
                if checkArray[len(checkArray)-1] == "[":
                    del checkArray[len(checkArray)-1]
                else:
                    return False
            if s[i] == "}":
                if checkArray[len(checkArray)-1] == "{":
                    del checkArray[len(checkArray)-1]
                else:
                    return False
    if len(checkArray) != 0:
        return False
    return True
    
s = "{[()]}"        
print(isValid(s))    
```
